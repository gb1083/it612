```python
%%writefile code.js

//A

const data = [5,12,3,20,11,9,30];

let loop = [];

for (let i = 0; i < data.length; i++) {
  const v = data[i];
  if (v > 10) {
    loop.push(v * 2)
  }
}
console.log(loop)

//pipeline

const pipeline = data
.filter(v => v > 10)
.map(v => v * 2)

console.log("Loop result: ", loop)
console.log("Pipeline result: ", pipeline)
console.log("Match:", JSON.stringify(loop) === JSON.stringify(pipeline));

// B
const numbers = [1,2,3,4,5,6]

const sumDoubledEvens = numbers
.filter(v => v % 2 === 0)
.map(v => v * 2)
.reduce((acc, v) => acc + v, 0)

console.log(" Sum of the douvled evens: ",sumDoubledEvens)

// C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

const under18 = users.some(user => user.age < 18);
const id3 = users.find(user => user.id === 3);


console.log("Is there any user under 18?", under18)
console.log("Find user with id 3", id3)

console.log("Users names:")
users.forEach(user => console.log(user.name))

//D

function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);

}
}
repeat(3, console.log);
repeat(4, i => console.log("Step " + i))


```
